## 1. Reading an image

```python
# Importing the OpenCV library 
import cv2 
print(cv2.__version__)
```

```python
# Reading the image using imread() function 
image = cv2.imread('.\\images\\road.jpg') 

# Extracting the height and width of an image 
h, w = image.shape[:2] 

# Displaying the height and width 
print("Height = {}, Width = {}".format(h, w))
```

### Show the image

```python
#The line above is necesary to show Matplotlib's plots inside a Jupyter Notebook
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#Show the image with matplotlib
plt.imshow(image)
plt.show()
```

```python
# The API documentation for cv2.waitKey()
cv2.imshow('img', image)
while(True):
    k = cv2.waitKey(33)
    if k == -1:  # if no key was pressed, -1 is returned
        continue
    else:
        break
        
cv2.destroyWindow('img')
```

```python
# load image using cv2....and do processing.
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

# as opencv loads in BGR format by default, we want to show it in RGB.
plt.show()
```

## 2. Extracting the RGB values of a pixel
Now we will focus on extracting the RGB values of an individual pixel.

__Note__  OpenCV arranges the channels in BGR order. So `the 0th value will correspond to Blue pixel and not Red`.

```python
# Extracting RGB values. 
# Here we have randomly chosen a pixel 
# by passing in 100, 100 for height and width. 
(B, G, R) = image[100, 100] 

# Displaying the pixel values 
print("R = {}, G = {}, B = {}".format(R, G, B)) 

# We can also pass the channel to extract 
# the value for a specific channel 
# B[0], G[1], R[2]
B = image[100, 100, 0] 
print("B = {}".format(B)) 
```

## 3. Extracting the Region of Interest (ROI)

```python
# We will calculate the region of interest 
# by slicing the pixels of the image 
roi = image[100 : 500, 200 : 700] 
plt.imshow(roi)
plt.show()
```

## 4. Resizing the Image

```python
# resize() function takes 2 parameters, 
# the image and the dimensions 
resize = cv2.resize(image, (800, 800)) 
plt.imshow(resize)
plt.show()
```

The problem with this approach is that the aspect ratio of the image is not maintained. So we need to do some extra work in order to maintain a proper aspect ratio.

```python
# Calculating the ratio 
ratio = 800 / w 

# Creating a tuple containing width and height 
dim = (800, int(h * ratio)) 

# Resizing the image 
resize_aspect = cv2.resize(image, dim) 
plt.imshow(resize_aspect)
plt.show()
```

```python
# Extracting the height and width of an image 
h1, w1 = resize_aspect.shape[:2] 

# Displaying the height and width 
print("Height = {}, Width = {}".format(h1, w1))
```

## 5. Rotating the Image

```python
# Calculating the center of the image 
center = (w // 2, h // 2) 

# Generating a rotation matrix 
matrix = cv2.getRotationMatrix2D(center, -60, 1.0) 

# Performing the affine transformation 
rotated = cv2.warpAffine(image, matrix, (w, h)) 
plt.imshow(rotated)
plt.show()
```

## 6. Drawing a Rectangle

```python
# We are copying the original image, 
# as it is an in-place operation. 
output = image.copy() 

# Using the rectangle() function to create a rectangle. 
#It takes in 5 arguments 
# - Image
# - Top-left corner co-ordinates
# - Bottom-right corner co-ordinates
# - Color (in BGR format)
# - Line width
rectangle = cv2.rectangle(output, (1500, 900), (600, 400), (255, 0, 0), 5) 
plt.imshow(rectangle)
plt.show()
```

## 7. Displaying text

```python
# Copying the original image 
output = image.copy() 

# Adding the text using putText() function 
#It takes in 7 arguments 
# - Image
# - Text to be displayed
# - Bottom-left corner co-ordinates, from where the text should start
# - Font
# - Font size
# - Color (BGR format)
# - Line width
text = cv2.putText(output, 'OpenCV Demo', (500, 550), cv2.FONT_HERSHEY_SIMPLEX, 4, (255, 0, 0), 5) 
plt.imshow(text)
plt.show()
```
