__cv2.imshow()__ method is used to display an image in a window. The window automatically fits to the image size


```
Syntax: cv2.imshow(window_name, image)
Parameters: 
    window_name: A string representing the name of the window in which image to be displayed. 
    image: It is the image that is to be displayed.
    Return Value: It doesnâ€™t returns anything. 
```

````python
# Importing the OpenCV library 
import cv2 
print(cv2.__version__)

#The line above is necesary to show Matplotlib's plots inside a Jupyter Notebook
from matplotlib import pyplot as plt
%matplotlib inline
````

````python
# Python program to explain cv2.imshow() method 

# path 
path = r'.\images\geeks14.png'

# Reading an image in default mode 
image = cv2.imread(path) 
````

````python
# Window name in which image is displayed 
window_name = 'Show Geek Image'

# Using cv2.imshow() method 
# Displaying the image 
cv2.imshow(window_name, image) 

#waits for user to press any key 
#(this is necessary to avoid Python kernel form crashing) 
cv2.waitKey(0) 

#closing all open windows 
cv2.destroyAllWindows() 
````

````python
#Show the image with matplotlib
plt.imshow(image)
plt.show()
````

## Showing in grascale mode

````python
# Reading an image in grayscale mode 
image = cv2.imread(path, 0) 

# Window name in which image is displayed 
window_name = 'Show Geek Image'

# Using cv2.imshow() method 
# Displaying the image 
cv2.imshow(window_name, image) 

# waits for user to press any key 
# (this is necessary to avoid Python kernel form crashing) 
cv2.waitKey(0) 

# closing all open windows 
cv2.destroyAllWindows()
````

````python
#Show the image with matplotlib
plt.imshow(image)
plt.show()
````
