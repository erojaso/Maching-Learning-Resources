# How to: Boxcox transform on train & test data
[Kaggle](https://www.kaggle.com/rtatman/boxcox-transform-on-train-test-data)

```python
# import modules
import numpy as np
from scipy import stats
from sklearn.model_selection import train_test_split
from pylab import rcParams

# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# generate non-normal data
original_data = np.random.exponential(size = 1000)

rcParams['figure.figsize'] = 15, 5
plt.plot(original_data)
plt.show()
```

```python
sns.distplot(original_data, kde=True)
plt.tight_layout()
```

```python
# transform training data & save lambda value
adjust_data,fitted_lambda = stats.boxcox(original_data)
fitted_lambda
```

```python
rcParams['figure.figsize'] = 15, 5
plt.plot(adjust_data)
plt.show()
```

```python
sns.distplot(adjust_data, kde=True)
plt.tight_layout()
```

```python
# split into testing & training data
train, test = train_test_split(original_data, shuffle=False)
print(len(original_data))
print(len(train))
print(len(test))
```

```python
# (optional) plot train & test
fig, ax = plt.subplots(1,2)
sns.distplot(train, ax=ax[0])
sns.distplot(test, ax=ax[1])
```

```python
# transform training data & save lambda value
train_data,fitted_lambda = stats.boxcox(train)
print(fitted_lambda)

# use lambda value to transform test data
test_data = stats.boxcox(test, fitted_lambda)
```
